```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

def permutation(n, r):
    if r >= 0 or r <= 0:
        return factorial(n) / factorial(n - r)   
    else:
        return 0    

def combination(n, r):
    return permutation(n, r) * 1 / factorial(r)

combination(5, 2)        
```
